```python
def pendiente(x,y):
    pendiente=(y[size(y)-1]-y[0])/(x[size(x)-1]-x[0])
    return pendiente
def deriva(x,y,a):
    for i in range (size(x)):
        y[i]=y[i]-(a*(x[i]-x[0]))
    return y
```

# free air

```python
data7 = loadtxt("Drift_GrupoH_ok.dat", usecols = (0,1,2))
p1 = data7[:,0] #gravedad
h1= data7[:,1] #tiempo
j1=data7[:,2] #altura
plt.title('Correccion de Free-air grupo H')
scatter(p1,h1)
```

```python
def freeAir(h1,j1):
    horizonte=j1[0]
    for i in range (1,size(j1)):
        if j1[i]>horizonte:
            h1[i]=h1[i]+abs(j1[i]-j1[0])*0.3086
        else:
            h1[i]=h1[i]-abs(j1[i]-j1[0])*0.3068
    return h1
        
```

```python
freeAir(h1,j1)
```
