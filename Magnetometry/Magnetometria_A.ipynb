```python
#Codigo que carga los datos del grupo H.
f=open("GrupoA/GrupoA_cuadrilla1.csv")
datos= csv.reader(f,delimiter=';')
x=[]
y=[]
z=[]
for row in datos:
    p=float(row[0])
    q=float(row[1])
    r=float(row[2])
    x.append(p)
    y.append(q)
    z.append(r)
    
f1=open("GrupoA/GrupoA_cuadrilla3.csv")
datos1= csv.reader(f1,delimiter=';')
x1=[]
y1=[]
z1=[]
for row in datos1:
    p1=float(row[0])
    q1=float(row[1])
    r1=float(row[2])
    x1.append(p1)
    y1.append(q1)
    z1.append(r1)
f2=open("GrupoA/GrupoA_cuadrilla2.csv")
datos2= csv.reader(f2,delimiter=';')
x2=[]
y2=[]
z2=[]
for row in datos2:
    p2=float(row[0])
    q2=float(row[1])
    r2=float(row[2])
    x2.append(p2)
    y2.append(q2)
    z2.append(r2)
```
